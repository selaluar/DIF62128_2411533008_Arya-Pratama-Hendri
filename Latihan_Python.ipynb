```python
x = 10
y = 3.14
name = "Arya"
is_active = True

print(x, type(x))
print(y, type(y))
print(name, type(name))
print(is_active, type(is_active))
```

```python
a = "25"
b = "2.5"

a_int = int(a)
b_float = float(b)

print(a_int, type(a_int))
print(b_float, type(b_float))

```

```python
a = 7
b = 3

print("a+b =", a+b)
print("a-b =", a-b)
print("a*b =", a*b)
print("a/b =", a/b)
print("a//b =", a//b)
print("a%b =", a%b)
print("a**b =", a**b)
```

```python
x = 10
y = 12

print(x == y)
print(x != y)
print(x < y)
print(x <= y)
print(x > y)
print(x >= y)
```

```python
score = 78

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "D"

print("Grade:", grade)
```

```python
text = "Belajar Python"

print("Length:", len(text))
print("Upper:", text.upper())
print("Lower:", text.lower())
print("Replace:", text.replace("Python", "Coding"))

print("First char:", text[0])
print("Last char:", text[-1])
print("Slice [0:7]:", text[0:7])
```
